<CENTER><h1>Simple pyROOT notebook example</h1></CENTER>

```python
import ROOT
```

```python
f = ROOT.TFile.Open("http://opendata.atlas.cern/release/samples/MC/mc_105986.ZZ.root")
# f = ROOT.TFile.Open("mc_105986.ZZ.root")
```

```python
c = ROOT.TCanvas("testCanvas","a first way to plot a variable",800,600)
```

```python
t = f.Get("mini")
```

```python
h = ROOT.TH1F("variable","Number of Lepton",4,0,8)
h2 = ROOT.TH1F("variable","Number of selected jets",4,0,8)
```

```python
for event in t:
    h.Fill(t.lep_n)
    h2.Fill(t.jet_n)
print "Done!"
```

```python
h.Draw()
h2.Draw("same")
```

```python
c.Draw()
```

```python
scale = h.Integral()
h.Scale(1/scale)
scale2 = h2.Integral()
h2.Scale(1/scale2)
h2.SetLineColor(2)
```

```python
h.Draw()
h2.Draw("same")
c.Draw()
```
