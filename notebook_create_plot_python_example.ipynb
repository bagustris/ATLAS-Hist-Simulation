<CENTER><h1>Simple pyROOT notebook example</h1></CENTER>

```python
import ROOT
```

```python
f = ROOT.TFile.Open("http://opendata.atlas.cern/release/samples/MC/mc_105986.ZZ.root")
# f = ROOT.TFile.Open("mc_105986.ZZ.root")
```

```python
c = ROOT.TCanvas("testCanvas","a first way to plot a variable",800,600)
```

```python
t = f.Get("mini")
```

```python
h = ROOT.TH1F("variable","Example plot: Number of Leptons",4,0,4)
```

```python
for event in t:
    h.Fill(t.lep_n)
    
print "Done!"
```

```python
h.Draw()
```

```python
c.Draw()
```

```python
scale = h.Integral()
h.Scale(1/scale)
```

```python
h.Draw()
c.Draw()
```
